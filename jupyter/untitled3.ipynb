```python
# MedianOfTwoSortedArrays
def Median(nums1, nums2):
    m, n = len(nums1), len(nums2)
    if m > n:
        nums1, nums2, m, n = nums2, nums1, n, m
    if n == 0:
        raise ValueError

    imin, imax, half_len = 0, m, (m+n+1)/2
    while imin <= imax:
        i = int((imin + imax) / 2)
        j = int(half_len - i) 
        if i < m and nums2[j-1] > nums1[i]:
            imin = i + 1
        elif i > 0 and nums1[i-1] > nums2[j]:
            imax = i - 1
        else:
            if i == 0: max_of_left = nums2[j-1]
            elif j == 0: max_of_left = nums1[i-1]
            else: max_of_left = max(nums1[i-1], nums2[j-1])
            
            if (m + n) % 2 == 1:
                return max_of_left
            
            if i == m: min_of_right=nums2[j]
            elif j == n: min_of_right=nums1[i]
            else: min_of_right=min(nums1[i], nums2[j])
                
            return (max_of_left + min_of_right) / 2.0
           
if __name__ == '__main__':
    l1 = [1, 2, 5]
    l2 = [3, 4]
    print(Median(l1, l2))
```

```python
# 集合
import itertools
def combine(n, k):
    return list(itertools.combinations(range(1,n+1), k))

# accepted
def combine1(n, k):
    if k == 0:
        return [[]]
    return [pre + [i] for i in range(k, n+1) for pre in combine1(i-1, k-1)]
    
def combine2(n, k):
    ans = []
    stack = []
    x = 1
    while True:
        l = len(stack)
        if l == k:
            ans.append(stack[:])
        if l == k or x > n - k +  + 1:
            if not stack:
                return ans
            x = stack.pop() + 1
        else:
            stack.append(x)
            x += 1
    
if __name__ == '__main__':
    n = 5
    k = 2
    print(combine2(n, k))
```

```python
# combine sum
def combination(l, target):
    l.sort()
    result = []
    dfs(l, 0, [], result, target)
    return result

def dfs(nums, start, path, result, target):
    if not target:
        result.append(path)
        return result
    for i in range(start, len(nums)):
        if nums[i]>target:
            break
#         i+1->i the number can be duplicate
        dfs(nums, i+1, path+[nums[i]], result, target-nums[i])
        
if __name__ == '__main__':
    l = [2,3,4,7]
    target = 7
    print(combination(l, target))  
```

```python
# 最长回文串
def longestPalindrome(s):
    if len(s)==0:
        return 0
    maxlen = 1
    start = 0
    for i in range(len(s)):
        if i-maxlen >= 1 and s[i-maxlen-1:i+1]==s[i-maxlen-1:i+1][::-1]:
            start = i-maxlen-1
            maxlen+=2
            continue
        if i-maxlen>=0 and s[i-maxlen:i+1]==s[i-maxlen:i+1][::-1]:
            start = i-maxlen
            maxlen+=1
    return s[start:start+maxlen]

if __name__=='__main__':
    s = 'abac'
    print(longestPalindrome(s))
```

```python
# 最长子串
class Solution(object):
    def lengthOfLongestSubSequence(self, s):
        n = 0
        l = []
        for i in s:
            if i not in l:
                l.append(i)
                n+=1
        print(l, n)
    
    def useSet(self, s):
        l = list(set(s))
        print(l)
        
    def useDict(self, s):
        l = {}.fromkeys(s).keys()
        print(l)
        
    def useDict_sort(self, s):
        l = list(set(s))
        l.sort(key=s.index)
        print(l, len(l))
        
    def Substring(self, s):
        start = maxlength = 0
        usedChar = {}
        
        for i in range(len(s)):
            if s[i] in usedChar and start <= usedChar[s[i]]:
                start = usedChar[s[i]] + 1
            else:
                maxlength = max(maxlength, i-start+1)
            usedChar[s[i]] = i
        print(maxlength)
    
if __name__ == '__main__':
    test = Solution()
    s = ['p', 'w', 'w', 'k', 'e', 'w']
    test.lengthOfLongestSubSequence(s)
    test.useSet(s)
    test.useDict(s)
    test.useDict_sort(s)
    test.Substring(s)
```

```python
# 找零问题
def coinChange(values, money):
    coinsUsed = {i:0 for i in range(money+1)}
    for cents in range(1, money+1):
        minCoins = cents
        for value in values:
            if value <= cents:
                temp = coinsUsed[cents - value] + 1
                if temp < minCoins:
                    minCoins = temp
        coinsUsed[cents] = minCoins
    print(cents, coinsUsed[cents])
        
if __name__ == '__main__':
    value = [25,21,10,5,1]
    money = 63
    coinChange(value, money)
```

```python
# 找哪些钱
def giveChange(need_change, currency_list, num_list, used_list):
    # need_change 需要找零的金额
    # currency_list 面值列表
    # num_list used_list 需要找零的最少货币数目
    for change in range(need_change+1):
        for currency in currency_list:
            if (change-currency) >= 0 and (num_list[change-currency]+1 <= num_list[change]):
                num_list[change] = num_list[change-currency]+1
                used_list[change] = currency # 记录消耗的货币
    return

def showChange(need_change, used_list):
    give_list = []
    while need_change > 0:
        give_list.append(used_list[need_change])
        need_change -= used_list[need_change]
    give_list.sort()
    return give_list

if __name__ == '__main__':
    need_change = 63
    currency_list = [1,5,10,21,25]
    num_list = list(range(need_change+1))
    used_list = list(range(need_change+1))
    giveChange(need_change, currency_list, num_list, used_list)
    print(num_list)
    print(used_list)
    print("{0} 需要 {1} 个货币来找零".format(need_change, num_list[need_change]))
    give_list = showChange(need_change, used_list)
    print(give_list)
```
