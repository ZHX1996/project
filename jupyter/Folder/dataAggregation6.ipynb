```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

# origindata1 = pd.Series([24,27,24,26,26,28,26,26,27,26,25,26,26,25,26,27,26,26,25,25,25,25,27,25])  # 24
origindata = pd.Series([29,28.6,26.5,24.8,24.9,25.5,26,25.8,25.6,24.9,25,24.8,24.8,24.3,24,23.9,24.2,
                        24.9,26.7,28.3,29.1,29.1,29.7,30.3,30.7,30.7,30.2,29.6,28.3,26.7,25.8,25.2,
                        24.8,24,24.3,22.9,23.2,23.2,22.4,21.5,21.3,23.2,26.1,27.7,28.5,29.3,29.9,
                        30.6,31.2,31,30.9,   29.8,29.3,28.9,27.8,27.1,27.1,26.7,26.1,26,25.9,25.8,25.6,24.9,
                         24.6,24.6,25.6,26.8,27.6,28.6,28.6,29.9,30.4,30.6,30.4,30.6,29.5,28.8,27.7,
                        27.1,26.4,25.9,25.4,25.1,24.7,24.1,24,23.8,23.1,23,23.9,25.5,27.2, 28.1,29.2,
                        30.3,30.6,31.3,31.1,31
                    ])
np.random.seed(1231)
noise1 = np.random.randn(100)
data1 = origindata+noise1
np.random.seed(1232)
noise2 = np.random.randn(100)
data2 = origindata+noise2
np.random.seed(1233)
noise3 = np.random.randn(100)
data3 = origindata+noise3
np.random.seed(1234)
noise4 = np.random.randn(100)
data4 = origindata+noise4
np.random.seed(1235)
noise5 = np.random.randn(100)
data5 = origindata+noise5
np.random.seed(1236)
noise6 = np.random.randn(100)
data6 = origindata+noise6

    
def kalman(z):
    n_iter = 100
    sz  = (n_iter,)
    # x = -0.37727
    # z = np.random.normal(x, 0.1, size=sz)
    Q = 1.2e-5
    xhat = np.zeros(sz)
    P = np.zeros(sz)
    xhatminus = np.zeros(sz)
    Pminus = np.zeros(sz)
    K = np.zeros(sz)
    R = 0.01**2
    xhat[0] = 0.0
    P[0] = 1.0

    for k in range(1,n_iter):
        xhatminus[k] = xhat[k-1]  
        Pminus[k] = P[k-1]+Q     

        K[k] = Pminus[k]/( Pminus[k]+R )
        xhat[k] = xhatminus[k]+K[k]*(z[k]-xhatminus[k]) 
        P[k] = (1-K[k])*Pminus[k]
    return xhat

var1 = kalman(data1)
var2 = kalman(data2)
var3 = kalman(data3)
var4 = kalman(data4)
var5 = kalman(data5)
var6 = kalman(data6)

res = []
for i in range(86):
    var = round((var1[i+14]+var2[i+14]+var3[i+14]+var4[i+14]+var5[i+14]+var6[i+14])/6,3)
    res.append(var)

np.set_printoptions(precision=3)
res = np.array(res)
print(res.tolist())
```
