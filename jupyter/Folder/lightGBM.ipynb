```python
# regression
import json
import lightgbm as lgb
import pandas as pd
from sklearn.metrics import roc_auc_score

df_train = pd.read_csv('F://houseworkData/lightGBM/regression_train.CSV', sep=',',header=None)
df_test = pd.read_csv('F://houseworkData/lightGBM/regression_test.CSV', sep=',',header=None)
y_train = df_train[0].values
y_test = df_test[0].values
X_train = df_train.drop(0, axis=1).values
X_test = df_test.drop(0, axis=1).values

lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test,y_test,reference=lgb_train)

params = {
    'task': 'train',
    'boosting_type':'gbdt',
    'objective':'binary',
    'metric':{'l2', 'auc'},
    'num_leaves':31,
    'learning_rate':0.05,
    'feature_fraction':0.8,
    'bagging_freq':5,
    'verbose':0
}

gbm = lgb.train(params,lgb_train,num_boost_round=20,valid_sets=lgb_eval,early_stopping_rounds=5)

y_pred = gbm.predict(X_test,num_iteration=gbm.best_iteration)
print(y_pred)
print('the roc of prediction is:',roc_auc_score(y_test,y_pred))
print('feature names:', gbm.feature_name())
print('feature importances:',list(gbm.feature_importance()))
```
