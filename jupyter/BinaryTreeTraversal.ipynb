```python
class Solution:
    # recursive
    def InorderTraversal1(self, root):
        res = []
        self.Inorderdfs(root, res)
        return res
    
    def Inorderdfs(self, root, res):
        if root:
            self.Inorderdfs(root.left, res)
            res.append(root.val)
            self.Inorderdfs(root.right, res)
 

    def PreorderTraversal1(self, root):
        res = []
        self.Preorderdfs(res, root)
        return res
    
    def Preorderdfs(self, res, root):
        if root:
            res.append(root.val)
            self.Preorderdfs(res, root.left)
            self.Preorderdfs(res, root.right)
            
    
    def PostorderTraversal1(self, root):
        res = []
        self.Postorderdfs(res, root)
        return res
    
    def Postorderdfs(self, res, root):
        if root:
            self.Postorderdfs(self, res, root.left)
            self.Postorderdfs(self, res, root.right)
            res.append(root.val)
               
            
    # iterative
    def InorderTraversal2(self, root):
        res, stack = [], []
        while True:
            while root:
                stack.append(root)
                root = root.left
            if not stack:
                return res
            node = stack.pop()
            res.append(node.val)
            root = node.right
            
    def PreorderTraversal2(self, root):
        res = []
        stack = [root]
        while stack:
            node = stack.pop()
            if node:
                res.append(node.val)
                stack.append(node.right)
                stack.append(node.left)
        return res
    
    # pre-order, right first
    def PostorderTraversal2(self, root):
        res = []
        stack = [root]
        while stack:
            node = stack.pop()
            if node:
                res.append(node.val)
                stack.append(node.left)
                stack.append(node.right)
        return res[::-1]
        
    def PostorderTraversal3(self, root):
        res, stack = [], [(root, False)]
        while stack:
            node, visited = stack.pop()
            if node:
                if visited:
                    res.append(node.val)
                else:
                    stack.append((node, True))
                    stack.append((node.right, False))
                    stack.append((node.left, False))
        return res
```
